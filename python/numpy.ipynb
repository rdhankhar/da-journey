```python
# Aggregate function

import numpy as np

array = np.array([[1,2,3,4],
                  [5,6,7,8],
                  [5,6,7,100]])

print(np.sum(array))
print(np.mean(array))
print(np.std(array))
print(np.var(array))
print(np.min(array))
print(np.max(array))
print(np.argmin(array))
print(np.argmax(array))
print(np.sum(array, axis = 1))
```

```python
# Filtering

import numpy as np

ages = np.array([[17,18,27,67,76,54,89,99,15,45],
                 [34,24,16,23,67,45,18,19,20,21]])

teenagers = ages[ ages < 18 ]
adults = ages[ (ages >= 18) & (ages < 65)]
adultss = ages[ (ages >= 18) | (ages < 65)]
even = ages[ages % 2 == 0]
odd = ages[ages % 2 != 0]
adult = np.where(ages >= 18 ,ages ,0)
adullt = np.where(ages % 2 == 0 ,ages ,-5)

print(teenagers)
print(adults)
print(adultss)
print(even)
print(odd)
print(adult)
print(adullt)
```

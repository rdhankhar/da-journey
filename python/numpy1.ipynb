```python
# Element wise airthmetic

array1 = np.array([1,2,3,4])
array2 = np.array([5,6,7,8])

print(array1 + array2)
print(array1 - array2)
print(array1 * array2)
print(array1 / array2)
print(array1 ** array2)
```

```python
# Aggregate function

array = np.array([[1,2,3,4],
                  [5,6,7,8],
                  [5,6,7,100]])

print(np.sum(array))
print(np.mean(array))
print(np.std(array))
print(np.var(array))
print(np.min(array))
print(np.max(array))
print(np.argmin(array))
print(np.argmax(array))
print(np.sum(array, axis = 1))
```

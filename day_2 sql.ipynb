```python
import pandas as pd 
from pandasql import sqldf

df = pd.read_csv(r"C:\Users\ASUS\DA Journey\train.csv")

pysql = lambda q : sqldf(q, globals() )

print("dataset shape = ", df.shape)
```

```python
result = pysql("""
Select Name , Fare , Pclass 
from df
where Fare > (SELECT AVG(Fare) from df)
order by Fare Desc
""")
result
```

```python
result = pysql("""
SELECT Pclass , avg_fare 
from (
select Pclass, Round ( AVG(Fare) ,2) as avg_fare from df
group by Pclass)
where avg_fare >50 
""")
result
```

```python
result = pysql("""
SELECT Name , Pclass , Fare 
FROM  df as outer_df 
where Fare > (SELECT  ROUND ( Avg(Fare), 2)from df 
as inner_df
WHERE inner_df.Pclass = outer_df.Pclass)

ORDER BY Pclass , Fare DESC 
""")
result 
               
```
